```python
import pandas as pd
url = 'beer.txt'
beer = pd.read_csv(url, sep=' ')
beer
```

```python
X=beer.drop('name',axis=1)
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,random_state=1)
km.fit(X)
```

```python
km.labels_
```

```python
beer['cluster']=km.labels_
beer.sort_values('cluster')
```

```python
center=beer.groupby('cluster').mean()
center
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.scatter(beer.calories,beer.alcohol,c=beer.cluster,s=50)
plt.scatter(center.calories,center.alcohol,c='black',marker='+',s=200)
plt.xlabel('calories')
plt.ylabel('alcohol')
```

```python
import numpy as np
colors = np.array(['red', 'green', 'blue', 'yellow'])
pd.scatter_matrix(X,c=colors[beer.cluster],figsize=(10,10),s=100)
```

```python
#Repeat with Scaled Data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
km=KMeans(n_clusters=3,random_state=1)
km.fit(X_scaled)
```

```python

beer['cluster'] = km.labels_
beer.sort_values('cluster')
```

```python
beer.groupby('cluster').mean()
```

```python
pd.scatter_matrix(X,c=colors[beer.cluster],figsize=(10,10),s=100)
```

```python
from sklearn import metrics
metrics.silhouette_score(X_scaled,km.labels_)
```

```python
k_range=range(2,20)
scores=[]
for k in k_range:
    km=KMeans(n_clusters=k,random_state=1)
    km.fit(X_scaled)
    scores.append(metrics.silhouette_score(X_scaled,km.labels_))


    
```

```python
plt.plot(k_range,scores)
plt.xlabel("Range")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

```python
km=KMeans(n_clusters=4,random_state=1)
km.fit(X_scaled)
beer['cluster']=km.labels_
beer.sort_values('cluster')
```
