```python
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split

import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
```

```python
cancer=load_breast_cancer()
X_train,X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,random_state=42,stratify=cancer.target)
logreg=LogisticRegression()
logreg.fit(X_train,y_train)
```

```python
print 'Accuracy of training subset {:.3f}'.format(logreg.score(X_train,y_train))
print 'Accuracy of testing subset {:.3f}'.format(logreg.score(X_test,y_test))
```

```python
#C->Control of strength
#lower c=>log_reg adjusts to majority data points
#Higher C==>Correct classification of data points
log_reg100=LogisticRegression(C=100)
log_reg100.fit(X_train,y_train)
print 'Accuracy of training subset{:.3f}'.format(log_reg100.score(X_train,y_train))
print 'Accuracy of testing subset{:.3f}'.format(log_reg100.score(X_test,y_test))
```

```python
#C->Control of strength
#lower c=>log_reg adjusts to majority data points
#Higher C==>Correct classification of data points
log_reg001=LogisticRegression(C=0.01)
log_reg001.fit(X_train,y_train)
print 'Accuracy of training subset{:.3f}'.format(log_reg001.score(X_train,y_train))
print 'Accuracy of testing subset{:.3f}'.format(log_reg001.score(X_test,y_test))
```

```python
plt.plot(logreg.coef_.T,'o',label='C=1')
plt.plot(log_reg001.coef_.T,'^',label='C=001')
plt.plot(log_reg100.coef_.T,'v',label='C=100')
plt.xticks(range(cancer.data.shape[1]),cancer.feature_names,rotation=90)
plt.xlabel('Coefficient index')
plt.ylabel('Coefficient Magnitude')
plt.legend()
```
